```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"/content/drive/MyDrive/Merge, Join, and Concatenate/Data1.csv")
df1
```

```python
df2 = pd.read_csv(r"/content/drive/MyDrive/Merge, Join, and Concatenate/Data2.csv")
df2
```

```python
df1.merge(df2, how = 'inner', on = ['FirstName', 'FellowshipID'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left',rsuffix = '_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left',rsuffix = '_Right', how = 'outer')
df4
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```
